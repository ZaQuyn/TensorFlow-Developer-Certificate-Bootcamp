**The dot product**

Matrix multiplication is also reffered to as the dot product

You can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`

```python
X, Y
```

```python
# Perform the dot product on X and Y (requires X or Y to be transposed)
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
# Check the value of Y, reshape Y and trnaspose Y
print("Normal Y:")
print(Y, "\n")

print("Y reshaped to (2, 3):")
print(tf.reshape(Y, shape=(2, 3)), "\n")

print("Y transposed:")
print(tf.transpose(Y))
```

```python
tf.matmul(X, tf.transpose(Y))
```

Generally, hen performing matrix multiplication on 2 tensors and one of the axes doesn't line up, u will transpose (rather than reshape) one of the tensors to satisfy the matrix multiplication rules.

### Aggregating tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values

```python
D = tf.constant([-7. -10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Let's go through the following forms of aggregation:
* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Creating a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0 ,100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the variance
tf.math.reduce_variance(tf.cast(E, dtype=tf.float16))
```

**Find the positional maximum and minimum**



```python
# Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Find the minimum using the positional minimum index
F[tf.argmin(F)]
```
